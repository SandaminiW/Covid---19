Drop column
'Sno','Time','ConfirmedIndianNational','ConfirmedForeignNational

```python
covid_df.drop(columns=['Sno','Time','ConfirmedIndianNational','ConfirmedForeignNational'],inplace=True, axis = 1)
```

```python
covid_df.head()
```

Date Time function

```python
covid_df['Date'] = pd.to_datetime(covid_df['Date'], format = '%Y-%m-%d')
```

```python
covid_df.head()
```

```python
# Active cases

covid_df['Active'] = covid_df['Confirmed'] - (covid_df['Cured'] + covid_df['Deaths'])
covid_df.tail()
```

Pivot Table

```python
statewise = pd.pivot_table(covid_df, values = ['Confirmed','Cured','Deaths'], index = 'State/UnionTerritory', aggfunc = max)
```

```python
statewise["Recovery Rate"] = statewise["Cured"]*100/statewise["Confirmed"]
```

```python
statewise["Mortality Rate"] = statewise["Deaths"]*100/statewise["Confirmed"]
```

```python
statewise = statewise.sort_values(by = "Confirmed", ascending = False)
```

```python
statewise.style.background_gradient(cmap = 'Reds')
```

Group by Function

```python
#  Top 10 active case states

top_10_active_cases = covid_df.groupby(by = 'State/UnionTerritory').max()[['Active', 'Date']].sort_values(by = ['Active'],ascending = False).reset_index()
```

Fig Function

```python
fig = plt.figure(figsize = (16,9))
```

```python
plt.title('Top 10 active cases in India', size = 25)
```

Bar plot

```python
ax = sns.barplot(data = top_10_active_cases.iloc[:10], x = 'State/UnionTerritory', y = 'Active')
```

```python
#  Top 10 active case states
top_10_active_cases = covid_df.groupby(by = 'State/UnionTerritory').max()[['Active', 'Date']].sort_values(by = ['Active'],ascending = False).reset_index()
fig= plt.figure(figsize = (16,9))
plt.title('Top 10 active cases in India', size = 25)
ax = sns.barplot(data = top_10_active_cases.iloc[:10], x = 'State/UnionTerritory', y = 'Active')

plt.xlabel('States')
plt.ylabel('Active')
plt.show()
```

```python
# Top states with highest deaths

top_10_deaths = covid_df.groupby(by = 'State/UnionTerritory').max()[['Deaths', 'Date']].sort_values(by = ['Deaths'],ascending = False).reset_index()
fig = plt.figure(figsize = (18,5))
plt.title('Top 10 states with highest deaths', size = 25)
ax = sns.barplot(data = top_10_deaths.iloc[:12], x = 'State/UnionTerritory', y = 'Deaths',)

plt.xlabel('States')
plt.ylabel('Deaths')
plt.show()
```
